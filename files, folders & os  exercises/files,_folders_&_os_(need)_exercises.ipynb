```python
import os
import glob
import shutil
import numpy as np
```

**Task 1**

```python
month_converter = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}
old_file_name = 'oct-08_singapore.txt'

#Generate new file name
month = old_file_name[0:3]
day = old_file_name[3:6]
the_rest = old_file_name[7:]
new_file_name = month_converter[month] + day + '-' + the_rest

#Rename
try: os.rename(old_file_name, new_file_name)
except: print('The file has been renamed!')
```

**Task 2**

```python
def rename_my_file(old_file_name): 
    month = old_file_name[0:3]
    day = old_file_name[3:6]
    the_rest = old_file_name[7:]
    new_file_name = month_converter[month] + day + '-' + the_rest
    return new_file_name
old_file_name = 'oct-10_singapore.txt'
new_file_name = rename_my_file(old_file_name)
```

**Task 3**

```python
list = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']
new_list = []
for old_file_name in list:
    new_file_name = rename_my_file(old_file_name)
    new_list.append(new_file_name)
    print(f'{old_file_name} ----> {new_file_name}')
```

**Task 4**

```python
old_paths = glob.glob('os-collaboration-exercise-data/*.txt', recursive = True)

for old_path in old_paths:
    dir_name, old_file_name = os.path.split(old_path)
    try:
        new_file_name = rename_my_file(old_file_name)
        new_path = dir_name + '\\' + new_file_name
        os.rename(old_path, new_path)
    except:
        print(f'The file: {old_file_name} has been renamed.')
    
```

**Task 5**

```python
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

try: os.mkdir('Cities')
except: print('The folder: Cities has been created.', '\n')
    
for city in cities:
    path = os.path.join('Cities', city)
    try:
        os.mkdir(path)
        print(f'Creating {path}')
    except:
        print(f'The folder: {city} has been created.')
```

**Task 6**

```python
cairo_file_paths = glob.glob('os-collaboration-exercise-data/*cai*', recursive = True)
cairo_destination = os.path.join('Cities', 'Cairo')

for cairo_file_path in cairo_file_paths:
    shutil.copy(cairo_file_path, cairo_destination)
    dir_name, file_name = os.path.split(cairo_file_path)
    print(f'The file: {file_name} has been copied.')
```

**Task 7**

```python
for index in range(len(cities)):
    #Generate path
    city = cities[index]
    lowercase_city = city.lower()
    file_paths = glob.glob(f'os-collaboration-exercise-data/*{lowercase_city}*', recursive = True)
    city_destination = os.path.join('Cities', city)
    
    #Copy files
    for file_path in file_paths:
        shutil.copy(file_path, city_destination)
        dir_name, file_name = os.path.split(file_path)
        print(f'The file: {file_name} has been copied.')
```
